```python
import numpy as np
import pandas as pd
import sklearn.datasets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading data from sklearn
data = sklearn.datasets.load_breast_cancer()

data1 = pd.DataFrame(data.data, columns = data.feature_names)
data1
```

```python
data1.isna().sum()
```

```python
#Loading Target variable into our datafram
data1['diagnosis'] = data.target
data1
```

```python
#1 --> Benign
#0 --> Malignant(Cancerous)
data1['diagnosis'].value_counts()
```

```python
data1.groupby('diagnosis').mean()
```

```python
#Splitting into X & Y
X = data1.drop('diagnosis', axis = 1)
Y = data1['diagnosis']
```

```python
#Train & Test splt
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Training the model
model = LogisticRegression()

model.fit(X_train, Y_train)
```

```python
#Evaluating the model
Y_pred = model.predict(X_test)
score = accuracy_score(Y_test, Y_pred)
print('Accuracy: {:.2f}%'.format(score * 100))
```
