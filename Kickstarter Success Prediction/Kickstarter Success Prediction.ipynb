```python
data = data.drop(data.query("state != 'failed' and state != 'successful'").index, axis = 0).reset_index(drop = True)
data
```

```python
data['state'].value_counts()
```

```python
#converting date to numerical values and append into DF
data['deadline-year'] = data['deadline'].apply(lambda x: np.float(x[0:4]))
data['deadline_month'] = data['deadline'].apply(lambda x: np.float(x[5:7]))

data['launched_year'] = data['launched'].apply(lambda x: np.float(x[0:4]))
data['launched_month'] = data['launched'].apply(lambda x: np.float(x[5:7]))

data = data.drop(['deadline','launched'], axis = 1)
data
```

```python
data['state'] = data['state'].apply(lambda x: 0 if x == 'failed' else 1)
data
```

```python
data['country'].unique()
```

```python
{column : list(data[column].unique()) for column in data.columns if data.dtypes[column] == 'object'}
```

```python
def onehot_encode(df, columns, prefixes):
    for col, prefix in zip(columns, prefixes):
        df = df.copy()
        dummy = pd.get_dummies(df[col])
        df = pd.concat([df, dummy], axis = 1)
        df = df.drop(col, axis = 1)
        
    return df
```

```python
data = onehot_encode(data, ['category', 'main_category', 'currency', 'country'], ['cat', 'main_cat', 'curr', 'ct'])
```

```python
data
```

```python
#Splitting
X = data.drop('state', axis = 1)
Y = data['state']
```

```python
X
```

```python
Y
```

```python
#Scaling
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

```python
#Splitting into train & test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
Y.mean()
```

```python
#For Imbalanced class we use class_weight
class_weights = class_weight.compute_class_weight(
    'balanced',
    Y_train.unique(),
    Y_train
)

class_weights = dict(enumerate(class_weights))
class_weights
```

```python
inputs = tf.keras.Input(shape = (221,))
x = tf.keras.layers.Dense(64, activation = 'relu')(inputs)
x = tf.keras.layers.Dense(64, activation = 'relu')(x)
outputs = tf.keras.layers.Dense(1, activation = 'sigmoid')(x)

model = tf.keras.Model(inputs = inputs, outputs = outputs)

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = [
        'accuracy',
        tf.keras.metrics.AUC(name = 'auc')
                ]
)

batch_size = 64
epochs = 100

history = model.fit(
    X_train,
    Y_train,
    validation_split=0.2,
    class_weight=class_weights,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True,
            verbose=1
        )
    ],
    verbose=2
)
```

```python
score = model.evaluate(X_test, Y_test)
score
```

```python
print('Accuracy: {:.2f}%'.format(score[1] * 100))
```
